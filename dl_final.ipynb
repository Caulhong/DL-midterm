# Cutout (Use sam to cut out mobile phone pictures with transparent background from pictures)

```python
from PIL import Image
import requests
from transformers import SamModel, SamProcessor

model = SamModel.from_pretrained("facebook/sam-vit-base").to("cuda")
processor = SamProcessor.from_pretrained("facebook/sam-vit-base")
image_path = "phone.png"
phone_init_image = Image.open(image_path).convert("RGB")
print(phone_init_image.size)
phone_init_image
```

```python
input_points = [[[400,400]]]
```

```python
inputs = processor(phone_init_image, input_points=input_points, return_tensors="pt").to("cuda")
outputs = model(**inputs)
masks = processor.image_processor.post_process_masks(outputs.pred_masks.cpu(), inputs["original_sizes"].cpu(), inputs["reshaped_input_sizes"].cpu())
scores = outputs.iou_scores
masks
```

```python
import torch
from PIL import Image
import numpy as np
masks_tensor = torch.stack(masks)
binary_masks = masks_tensor.int()
foregournd_mask_image = binary_masks[0][0][0].numpy() * 255  # Convert to numpy array and scale to 255 for image representation
foregournd_mask_image = Image.fromarray(foregournd_mask_image.astype(np.uint8))
foregournd_mask_image
```

```python
foregournd_mask_image.save("foregournd_mask_image.png")
```

```python
from PIL import Image
import numpy as np
mask_array = np.array(foregournd_mask_image)
mask_array = np.where(mask_array > 128, 255, 0).astype(np.uint8)
original_array = np.array(phone_init_image)
output_array = np.zeros((original_array.shape[0], original_array.shape[1], 4), dtype=np.uint8)
output_array[:, :, :3] = original_array[:, :, :3]
output_array[:, :, 3] = mask_array
foreground_noback_image = Image.fromarray(output_array, 'RGBA')
print(foreground_noback_image.size)
foreground_noback_image
```

```python
foreground_noback_image.save("foreground_noback_image.png")
```

# Extract boundaries

```python
from PIL import Image, ImageFilter, ImageChops


background_image = Image.open("dlbackground_image.png").convert("RGBA")
foreground_noback_image = Image.open("foreground_noback_image.png").convert("RGBA")
foreground_noback_image = foreground_noback_image.transpose(Image.FLIP_LEFT_RIGHT)

new_size = (420, 420)
foreground = foreground_noback_image.resize(new_size, Image.LANCZOS)
rotation_angle = -3
foreground = foreground.rotate(rotation_angle, expand=True)
position = (630, 180)

transparent = Image.new('RGBA', background_image.size, (0, 0, 0, 0))
transparent.paste(foreground, position, foreground)
foreground_with_background = Image.alpha_composite(background_image, transparent)
print(foreground_with_background.size)
foreground_with_background.save("foreground_with_background.png")

mask = foreground.split()[3]

result_image = Image.new("RGBA", background_image.size, (0, 0, 0, 255))

dilated = mask.filter(ImageFilter.MaxFilter(11))


eroded = mask.filter(ImageFilter.MinFilter(7))
outline = ImageChops.difference(dilated, eroded)

result_image.paste((255, 255, 255, 255), position, outline)

result_image.save("dlforeground_outline_mask.png")
result_image
```

# fix borders

```python
import torch
from torchvision import transforms

from pipeline_stable_diffusion_3_inpaint import StableDiffusion3InpaintPipeline
from diffusers.utils import load_image
pipe= StableDiffusion3InpaintPipeline.from_pretrained(
    "stabilityai/stable-diffusion-3-medium-diffusers",
    torch_dtype=torch.float16,
).to("cuda")
```

```python
prompt = "a phone is holded by three red fingers, missing area is the edge of phone, Inpaint the missing area with a smooth transitions"
source = Image.open("dlinpaint_source.png").convert("RGB")
mask = Image.open("dlforeground_outline_mask.png").convert("RGB")

result = pipe(
    prompt=prompt,
    image=source,
    mask_image=mask,
    height=1024,
    width=1024,
    num_inference_steps=50,
    guidance_scale=7.0,
    strength=1
).images[0]
result
```
